```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('/Dumbbell Plot_NFHS.csv')
df
```

```python
ordered_df = df.sort_values(by='NFHS4')
my_range=range(1,len(df.index)+1)
my_range
```

```python
ordered_df = df.sort_values(by='NFHS4')
my_range=range(1,len(df.index)+1)
my_range

plt.hlines(y=my_range, xmin=ordered_df['NFHS4'], xmax=ordered_df['NFHS5'], color='black', alpha=0.8)
plt.scatter(ordered_df['NFHS4'], my_range, color='blue', alpha=1, label='NFHS4' )
plt.scatter(ordered_df['NFHS5'], my_range, color='red', alpha=1 , label='NFHS5')
plt.legend()
plt.title("NFHS4 vs NFHS5 Data ", loc='center')
plt.xlabel('Percentage')
plt.ylabel('Parameters')
plt.yticks(my_range, ordered_df['Parameters '])
plt.grid()
plt.rcParams["figure.figsize"] = (5, 0.5)
```
